```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA Campaign Promotion

```python
file_path = r"C:\Users\tanap\OneDrive\เดสก์ท็อป\MADT3-NIDA\(2-2566) MADT 6004 Applied Data Analytics for Business\Model\Collaborative Filtering\credit_card_campaigns_data_v2.csv"
camp_df = pd.read_csv(file_path)
camp_df.head()
```

```python
summary_stats = camp_df.describe()
summary_stats
```

```python
missing_values = camp_df.isnull().sum().sort_values(ascending=False)
missing_values.head()
```

```python
camp_df['total_promotions'] = camp_df.iloc[:, 1:].sum(axis=1)
promotion_popularity = camp_df.iloc[:, 1:-1].sum().sort_values(ascending=False)
max_promotions = int(camp_df['total_promotions'].max())
camp_df[['total_promotions']].astype('int').sort_values(by='total_promotions', ascending=False).head()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 4))
sns.histplot(camp_df['total_promotions'], bins=range(0, max_promotions + 2), kde=False, color='skyblue')
plt.xlabel('Number of Promotions Participated')
plt.ylabel('Number of Customers')
plt.title('Distribution of Promotion Participation by Customers')
plt.show()

plt.figure(figsize=(10, 4))
sns.barplot(x=promotion_popularity.index, y=promotion_popularity.values, hue=promotion_popularity.index, palette='viridis', dodge=False)
plt.xticks(rotation=90)
plt.xlabel('Promotion')
plt.ylabel('Number of Participants')
plt.title('Popularity of Each Promotion')
plt.legend([],[], frameon=False)
plt.show()

plt.figure(figsize=(10, 5))
sns.heatmap(camp_df.iloc[:, 1:-1].transpose(), cmap='viridis', cbar=True)
plt.xlabel('Customer ID')
plt.ylabel('Promotion')
plt.title('Heatmap of Promotion Participation')
plt.show()
```

```python
sns.countplot(data=cust_cc_df.drop_duplicates(subset=['Customer_ID']), x='Gender', color='#4285f4')
plt.title('Gender Distribution')
plt.show()

sns.countplot(data=cust_cc_df.drop_duplicates(subset=['Customer_ID']), x='Marital_Status', color='#4285f4')
plt.title('Marital Status Distribution')
plt.show()

sns.countplot(data=cust_cc_df.drop_duplicates(subset=['Customer_ID']), x='Target', color='#4285f4')
plt.title('Target Distribution')
plt.show()

sns.countplot(data=cust_cc_df.drop_duplicates(subset=['Customer_ID']), x='Spend_Category', color='#4285f4')
plt.title('Tenure Distribution')
plt.show()
```

```python
unique_customers = cust_cc_df.drop_duplicates(subset=['Customer_ID'])

spend_target = unique_customers.groupby(['Spend_Category', 'Target']).size().unstack().fillna(0)
grand_total = spend_target.values.sum()
spend_target_percent = (spend_target / grand_total) * 100

custom_palette = sns.color_palette(["#aec7e8", "#1f77b4"])

plt.figure(figsize=(9, 5))
ax = spend_target_percent.plot(kind='bar', stacked=True, color=custom_palette, figsize=(9, 5))

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}%', 
                    (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.), 
                    ha='center', va='center', fontsize=10, color='black')

plt.title('Spend Category by Target (Percentage of Grand Total)')
plt.xlabel('Spend Category')
plt.ylabel('Percentage')
plt.show()
```
