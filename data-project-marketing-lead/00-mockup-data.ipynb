Mockup User Particication with Campaign Promotion for Each pro_restrnt

```python
np.random.seed(42)

num_users = 1000
num_pro_rests = 50

cust_ids = ["6600" + str(i+1).zfill(4) for i in range(num_users)]
pro_rest_campaigns = ["pro_restrnt" + str(i+1).zfill(2) for i in range(num_pro_rests)]
participation_matrix = np.zeros((num_users, num_pro_rests))
popularity = np.random.rand(num_pro_rests) ** 2
customer_activity = np.random.rand(num_users) ** 2

for i in range(num_users):
    for j in range(num_pro_rests):
        if np.random.rand() < (popularity[j] * customer_activity[i]):
            participation_matrix[i, j] = 1

df = pd.DataFrame(participation_matrix, columns=pro_rest_campaigns)
df.insert(0, 'cust_id', cust_ids)
df.head()
```

Mockup for Propensity to Use Model

```python
import pandas as pd
import numpy as np

num_customers = 1000
num_months = 12

np.random.seed(42)

date_range = pd.date_range(start='2023-01-01', periods=num_months, freq='M') #testt
customer_ids = ["6600" + str(i+1).zfill(4) for i in range(num_customers)]
data_list = []

for customer_id in customer_ids:

    age = np.random.choice(np.arange(18, 70))
    gender = np.random.choice(['Male', 'Female'], p=[0.48, 0.52])
    income = np.random.choice([15000, 25000, 35000, 45000, 55000, 65000, 75000, 85000, 95000, 105000], p=[0.2, 0.25, 0.2, 0.15, 0.1, 0.05, 0.02, 0.01, 0.01, 0.01])
    marital_status = np.random.choice(['Single', 'Married', 'Divorced', 'Widowed'], p=[0.4, 0.5, 0.05, 0.05])
    num_dependents = np.random.choice([0, 1, 2, 3, 4, 5], p=[0.4, 0.3, 0.15, 0.1, 0.04, 0.01])
    days_as_customer = np.random.randint(30, 3650)
    total_credit_limit = np.random.uniform(5000, 500000)
    
    for date in date_range:

        avg_monthly_spend = np.random.uniform(200, 30000)
        num_transactions_per_month = np.random.randint(1, 50)
        num_merchants_per_month = np.random.randint(1, 20)
        spend_category = np.random.choice(['Groceries', 'Entertainment', 'Travel', 'Dining', 'Health', 'Education'], p=[0.1, 0.2, 0.1, 0.3, 0.2, 0.1])
        credit_limit_utilization = np.random.uniform(0.1, 1.0)
        
        last_month_spend = avg_monthly_spend * np.random.uniform(1, total_credit_limit/avg_monthly_spend)
        last_6_months_spend = avg_monthly_spend * np.random.uniform(3, 6)
        last_12_months_spend = avg_monthly_spend * np.random.uniform(6, 12)
        remaining_credit_limit = total_credit_limit * (1 - credit_limit_utilization)
        num_overdue_payments_12_months = np.random.choice([0, 1, 2, 3, 4], p=[0.7, 0.2, 0.05, 0.03, 0.02])
        num_payments_in_full_12_months = np.random.randint(5, 12)
        num_late_payments_12_months = np.random.choice([0, 1, 2], p=[0.8, 0.15, 0.05])
        avg_payment_amount = last_month_spend * np.random.uniform(0.7, 1.2)
        days_since_last_transaction = np.random.randint(0, 30)
        
        high_spend = avg_monthly_spend >= 15000
        high_frequency = num_transactions_per_month >= 5
        high_utilization = credit_limit_utilization >= 0.55
        good_payment_history = num_payments_in_full_12_months / 12 >= 0.30
        dinner = spend_category in ['Dining', 'Travel']

        interaction_term = high_spend and high_frequency and high_utilization and good_payment_history and dinner
        base_prob = 0.05
        if interaction_term:
            response_prob = base_prob + 0.6
        else:
            response_prob = base_prob

        target = np.random.choice([0, 1], p=[1-response_prob, response_prob])

        data_list.append({
            'Customer_ID': customer_id,
            'Feature_Date': date,
            'Age': age,
            'Gender': gender,
            'Income': income,
            'Marital_Status': marital_status,
            'Number_of_Dependents': num_dependents,
            'Days_as_Customer': days_as_customer,
            'Total_Credit_Limit': total_credit_limit,
            'Average_Monthly_Spend': avg_monthly_spend,
            'Number_of_Transactions_Per_Month': num_transactions_per_month,
            'Number_of_Merchants_Per_Month': num_merchants_per_month,
            'Spend_Category': spend_category,
            'Credit_Limit_Utilization': credit_limit_utilization,
            'Number_of_Overdue_Payments_12_Months': np.random.choice([0, 1, 2, 3, 4], p=[0.7, 0.2, 0.05, 0.03, 0.02]),
            'Number_of_Payments_in_Full_12_Months': np.random.randint(5, 12),
            'Number_of_Late_Payments_12_Months': np.random.choice([0, 1, 2], p=[0.8, 0.15, 0.05]),
            'Average_Payment_Amt': avg_monthly_spend * np.random.uniform(0.7, 1.2),
            'Days_Since_Last_Transaction': np.random.randint(0, 30),
            'Avg_Spend_Per_Transaction': avg_monthly_spend / num_transactions_per_month,
            'High_Value_Transactions': np.sum(np.random.uniform(0, 1, num_transactions_per_month) > 0.9),
            'Num_Spend_Categories': np.random.randint(1, 6),
            'Avg_Days_Between_Transactions': 30 / num_transactions_per_month,
            'Times_Credit_Limit_Exceeded': np.random.randint(0, 3) if credit_limit_utilization > 0.8 else 0,
            'Previous_Month_Payment': avg_monthly_spend * np.random.uniform(0.5, 1.5),
            'Times_Payment_Missed': np.random.randint(0, 3) if np.random.uniform(0, 1) > 0.75 else 0,
            'Target': target
        })

data = pd.DataFrame(data_list)
data['Feature_Date'] = pd.to_datetime(data['Feature_Date'])
data.sort_values(by=['Customer_ID', 'Feature_Date'], inplace=True)
data['Target'] = data.groupby('Customer_ID')['Target'].shift(-2)
data['Target_Date'] = data['Feature_Date'] + pd.DateOffset(months=2)
data.dropna(subset=['Target'], inplace=True)
data.reset_index(drop=True, inplace=True)
print(data['Target'].value_counts())
data.head()
```
